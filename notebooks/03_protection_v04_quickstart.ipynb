Build the confidence sector, run one paired trial, and inspect the generated report. All quantities are normalized defensive receive metrics.


````python
from pathlib import Path
import sys, yaml
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))
from hpm_platform.workflows.protection_v04 import (
    _array_from_config, _sector_from_config, _run_one
)
config = yaml.safe_load((ROOT / 'configs/protection_v04.yaml').read_text())
array = _array_from_config(config)
sector = _sector_from_config(array, config)
array.n_elements, sector.n_grid_points, sector.half_width_deg
````
